```python
#$ALL #1
movies = db.movies.find({"genres":{ "$all":["Adventure", "Romance"]}}).limit(5)
for movie in movies:
    pprint(movie)
```

```python
#$ALL #2
movies = db.movies.find({"$and": [{"genres": "Adventure"}, {"cast": "Tom Hanks"}]}).limit(5)
for movie in movies:
    pprint(movie)
```

```python
#Aggregation #1
avgawa = [{"$group": {"_id": None ,"average_awards": {"$avg": "$awards.wins"}}}]
average_awards = db.movies.aggregate(avgawa)
for result in average_awards:
    pprint(result)
    

award_num = str(result['average_awards'])
    
"The average number of awards won by a movie is: "+ award_num
```

```python
# Aggregation #2
maxawa = [{"$group": {"_id": None, "most_awards": {"$max": "$awards.wins"}}}]
most_awards = db.movies.aggregate(maxawa)
for result in most_awards:
    pprint(result)
    
award_max = str(result['most_awards'])
"The highest number of awards won by a movie is: "+ award_max
```
